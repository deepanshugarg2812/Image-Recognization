```python
##The function return the image as array and the labels assigned with that if data is for training
def load_dataset(folder_name,isTrain):
    image_data = []
    if isTrain==True:
        labels = []
        dict_for_labels = {}
        cnt = 0
        for f in os.listdir(folder_name):
            dict_for_labels[f] = cnt
            print("Loading folder data ",f)
            p = os.path.join(folder_name,f)
            print(p)
            for ix in os.listdir(p):
                img = image.load_img(os.path.join(p,ix),target_size=(32,32))
                img = image.img_to_array(img)
                image_data.append(img)
                labels.append(dict_for_labels[f])
            cnt += 1
        return image_data,labels
    else:
        for f in os.listdir(folder_name):
            print("Loading folder data ",f)
            p = folder_name+f
            for ix in os.listdir(p):
                img = image.load_img(os.path.join(p,ix),target_size=(32,32))
                img = image.img_to_array(img)
                image_data.append(img)
        return image_data
```

```python
## To make the training of the model easy and to classifiy the image in one to many we create a dictonary
def ClassWiseData(x,y):
    data = {}
    for i in range(len(np.unique(y))):
        data[i] = []
    for i in range(x.shape[0]):
        data[y[i]].append(x[i])
    for k in data.keys():
        data[k] = np.array(data[k])
    return data
data = ClassWiseData(image_data,labels)
print(data[0].shape)
```

### Algorithm for classification

```python
class SVM:
    """Class by deepanshu garg"""
    def __init__(self,C):
        self.C = C
        self.W = 0
        self.B = 0
    def HingeLoss(self,W,B,X,Y):
        loss = 0.0
        loss += 0.5*np.dot(W,W.T)
        m = X.shape[0]
        c = self.C
        for i in range(m):
            ti = Y[i]*(np.dot(W,X[i].T) + B)
            loss += c*max(0,1-ti)
        return loss
    def fit(self,X,Y,learning_rate=0.001,batch_size=100,maxItr=300):
        no_of_features = X.shape[1]
        m = X.shape[0]
        n = learning_rate
        Weights = np.zeros((1,no_of_features))
        bias = 0
        c = self.C
        losses = []
        for i in range(maxItr):
            err = self.HingeLoss(Weights,bias,X,Y)
            losses.append(err)
            ids = np.arange(m)
            np.random.shuffle(ids)
            for batch_start in range(0,m,batch_size):
                grad_w = 0
                grad_b = 0
                for j in range(batch_start,batch_start+batch_size):
                    if j < m:
                        i = ids[j]
                        ti =  Y[i]*(np.dot(Weights,X[i].T)+bias)
                        if ti>1:
                            grad_w += 0
                            grad_b += 0
                        else:
                            grad_w += c*Y[i]*X[i]
                            grad_b += c*Y[i]
                Weights = Weights - n*Weights + n*grad_w
                bias = bias + n*grad_b
        self.W = Weights
        self.b = bias
        return Weights,bias,losses
```

As we are using the svm algo so we can do many vs one classification based on the one vs one classification so we need to combine any two classes at any one time

```python
def getDataPairForSVM(d1,d2):
    l1,l2 = d1.shape[0],d2.shape[0]
    samples = l1+l2
    features = d1.shape[1]
    data_pair = np.zeros((samples,features))
    data_labels = np.zeros((samples,))
    data_pair[:l1,:] = d1
    data_pair[l1:,:] = d2
    data_labels[:l1] = -1
    data_labels[l1:] = +1
    return data_pair,data_labels
```

### Training on data

```python
def trainSVMs(x,y):
    CLASSES = len(np.unique(y))
    svm_classifiers = {}
    mySVM  = SVM(1)
    for i in range(CLASSES):
        svm_classifiers[i] = {}
        for j in range(i+1,CLASSES):
            xpair,ypair = getDataPairForSVM(data[i],data[j])
            wts,b,loss = mySVM.fit(xpair,ypair,learning_rate=0.00001,maxItr=1000)
            svm_classifiers[i][j] = (wts,b)
            loss = np.array(loss)
            loss = loss.reshape((-1,1))
            plt.plot(loss)
            plt.show()
            
    
    return svm_classifiers
```

```python
svm_classifiers = trainSVMs(image_data,labels)
```

### Prediction

```python
def binaryPredict(x,w,b):
    z  = np.dot(x,w.T) + b
    if z>=0:
        return 1
    else:
        return -1
```

```python
def predict(x):
    CLASSES = 4
    count = np.zeros((CLASSES,))
    
    for i in range(CLASSES):
        for j in range(i+1,CLASSES):
            w,b = svm_classifiers[i][j]
            #Take a majority prediction 
            z = binaryPredict(x,w,b)
            
            if(z==1):
                count[j] += 1
            else:
                count[i] += 1
    
    final_prediction = np.argmax(count)
    #print(count)
    return final_prediction
```

```python
## Sample test
print(predict(image_data[120]))
print(labels[0])
```

```python
def accuracy(x,y):
    
    count = 0
    for i in range(x.shape[0]):
        prediction = predict(x[i])
        if(prediction==y[i]):
            count += 1
            
    return count/x.shape[0]
```
